**Q1.**

1. The paper is about data tidying and this can make the data cleaning process, and in turn data analysis much easier.
2. The tidy data standard provides a way to organize data values within a dataset, and by having a standard the process of cleaning is made easier since you don't have to start from scratch every time.
3. The sentence "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." means that every dataset may have some part of it that is considered messy or unnecessary or confusing etc but that "messy" part is not always the same between every dataset. The sentence "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general." means that it is difficult to strictly define variables and observations because of how what can be viewed as one or the other can change depending on context/scenario or even by looking at what other variables and observations there are in the dataset.
4. Wickham defines values as what datasets are comprised of and they are usually either numbers or strings (quantitave or qualitative). She defines variables as all values that measure the same underlying attribute across units and observations as all values measured on the same unit across attributes.
5. Tidy data is defined as a standard way of mapping the meaning of a dataset to its structure. In tidy data each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
6. The 5 most common problems with messy data sets are: column headers are values and not variable names, multiple variables stored in one column, variables are stored in both rows and columns, multiple types of observational units are stored in the same table, a single observational unit is stored in multiple tables. The data in Table 4 are messy because the column headers are values (in this case income) instead of variables. Melting a dataset refers to turning the columns of a dataset into rows.
7. Table 11 has a column, element, which contains variable names and with a different column for each date of the month with each column being mostly empty. Table 12 is tidy and molten because the separate columns for dates have been condensed into one and the element column has been split into two columns based on the variables that were stored there.
8. The "chicken-and-egg" problem is "if tidy data is only as useful as the tools that work with it, then tidy tools will be inextricably linked to tidy data (Wickham 20)." Wickham hopes that others will build on her framework for tidy data to make even better storage strategies and tools.

```python
imposed_sentence = df["ImposedSentenceAllChargeInContactEvent"]
sentence_type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

imposed_sentence = pd.to_numeric(imposed_sentence,errors='coerce')
imposed_sentence_null = imposed_sentence.isnull()
print("null vals:", np.sum(imposed_sentence_null))

print(pd.crosstab(imposed_sentence_null, sentence_type),'\n')

imposed_sentence = imposed_sentence.mask(sentence_type == 9,np.nan)

imposed_sentence = pd.to_numeric(imposed_sentence,errors='coerce')
imposed_sentence_null = imposed_sentence.isnull()
print("null vals:", np.sum(imposed_sentence_null))

print(pd.crosstab(imposed_sentence_null, sentence_type),'\n')

df["ImposedSentenceAllChargeInContactEvent"] = imposed_sentence
```
